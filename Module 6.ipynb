```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('week6.csv')
print(data.shape)
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
data = data.pct_change()*100
```

```python
data.fillna(0,inplace=True)
```

```python
data
```

# 6.2

```python
data.head()
```

```python
def risk_and_volatility(data):
    data['Weighted Return'] = 0.0
    weight = 0.03
    for i in data.index:
        data['Weighted Return'][i] = (weight * np.sum(data.ix[i]) )
    annual_return = np.mean(data['Weighted Return'])*252
    volatility = np.std(data['Weighted Return'])* np.sqrt(252)
    print(f'The Overall Annual Return of the Portfolio is: {annual_return} %\nThe Overall Volatility of the Portfolio is: {volatility}')
risk_and_volatility(data)
```

```python
expected_return_risk = pd.DataFrame()
```

```python
data_arr = np.array(data.iloc[:,:-1])
print(data_arr.shape)
```

```python
return_each_stock = (np.mean(data_arr,axis=0)) * 252
print(return_each_stock.shape)
print(type(return_each_stock))
return_each_stock = return_each_stock.reshape(-1,1)
print(return_each_stock.shape)
```

```python
col = []
for i in range(1,31):
    col.append(f'Annual Return {str(i)}')
col = np.array(col)
```

```python
expected_return_risk = pd.DataFrame(return_each_stock,index=col,columns= ['Expected Return %'])
expected_return_risk
```

```python
risk_each_stock = (np.std(data_arr,axis=0)) * np.sqrt(252)
print(risk_each_stock.shape)
print(type(risk_each_stock))
risk_each_stock = risk_each_stock.reshape(-1,1)
print(risk_each_stock.shape)
```

```python
expected_return_risk['Volatility % '] = risk_each_stock
```

```python
expected_return_risk
```

# 6.3

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(expected_return_risk)
```

```python
plt.scatter(expected_return_risk.iloc[:,0],expected_return_risk.iloc[:,1], c=kmeans.labels_, cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt


data = np.asarray([np.asarray(expected_return_risk['Expected Return %']), np.asarray(expected_return_risk['Volatility % '])]).T
X = data
distortions = []
K = range(1,15)
for k in K:
    KmeanModel = KMeans(n_clusters=k).fit(X)
    KmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, KmeanModel.cluster_centers_, 'euclidean'),axis=1)) // X.shape[0])
    
    
#plot
plt.figure(figsize=(20,10))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortions')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
Y_pred = kmeans.predict(expected_return_risk)
```

```python
Y_pred
```
