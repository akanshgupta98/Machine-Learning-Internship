```python

def average_price_bt_days(N):
    return(np.mean(data_copy['Close Price'][-N:]))
    

```

```python
def per_change_bt_days(N):
    diff = data_copy['Close Price'][data_copy.index[-1]] - data_copy['Close Price'][data_copy.index[-N]]
    return((diff/data_copy['Close Price'][data_copy.index[-N]])*100)
```

```python
print("The average price change since N last days")
print("Last 7 days:", average_price_bt_days(7))
print("Last 14 days:",average_price_bt_days(14))
print("Last 1 month:",average_price_bt_days(30))
print("Last 3 months:",average_price_bt_days(90))
print("Last 6 months:",average_price_bt_days(180))
print("Last 1 year:",average_price_bt_days(365))

```

```python
print("The percentage Change since N last days:")
print("Last 7 days:", per_change_bt_days(7))
print("Last 14 days:",per_change_bt_days(14))
print("Last 1 month:",per_change_bt_days(30))
print("Last 3 months:",per_change_bt_days(90))
print("Last 6 months:",per_change_bt_days(180))
print("Last 1 year:",per_change_bt_days(365))

```

# 1.7

```python
data_copy['Trend'] = 'lose'
for i in data_copy.index:
    if( -0.5 <= data_copy['Day_perc_change'][i] <0.5):
        data_copy['Trend'][i] = 'Slight Change'
    elif(0.5 <= data_copy['Day_perc_change'][i] <1):
        data_copy['Trend'][i] = 'Slight Positive'
    elif(-1 <= data_copy['Day_perc_change'][i] <-0.5):
        data_copy['Trend'][i] = 'Slight Negative'
    elif(1 <= data_copy['Day_perc_change'][i] <3):
        data_copy['Trend'][i] = 'Positive'
    elif(-3 <= data_copy['Day_perc_change'][i] <-1):
        data_copy['Trend'][i] = 'Negative'
    elif(3 <= data_copy['Day_perc_change'][i] <7):
        data_copy['Trend'][i] = 'Among Top Gainers'
    elif(-7 <= data_copy['Day_perc_change'][i] <-3):
        data_copy['Trend'][i] = 'Among Top Losers'
    elif(data_copy['Day_perc_change'][i] > 7):
        data_copy['Trend'][i] = 'Bull Run'
    elif(data_copy['Day_perc_change'][i] < -7):
        data_copy['Trend'][i] = 'Bear Drop'
```
