```python
g = {}
g = data_copy.groupby('Trend').size()
```

```python
g['Among Top Gainers']
```

```python
g = dict(g)
```

```python
plt.figure(figsize=(10,10))
plt.pie(g.values(),labels=g.keys(),autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
g = data_copy.groupby('Trend')
```

```python
mean_values = []
median_values = []
keys= [] 
for key,df_key in g:
    keys.append(key)
    mean_values.append(df_key['Total Traded Quantity'].mean())
    median_values.append(df_key['Total Traded Quantity'].median())
   
```

```python
print(mean_values)
print(keys)
```

```python

plt.figure(figsize = (15,5))
plt.bar(keys,mean_values)
plt.title('Mean')
plt.show()
plt.figure(figsize = (15,5))
plt.bar(keys,median_values,color='r')
plt.title('Median')

plt.show()
```
